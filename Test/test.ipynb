```python
#Додав морфологічну обробку
#Оптимізує для зображення 2
from __future__ import print_function
import cv2 as cv
import numpy as np

img = cv.imread('6.jpg')
img = cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_AREA)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold = 150
canny = cv.Canny(gray, threshold, threshold * 2)
contours, _ = cv.findContours(canny, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

contours_poly = [cv.approxPolyDP(c, 3, True) for c in contours]
centers = [cv.minEnclosingCircle(p)[0] for p in contours_poly]
radii = [cv.minEnclosingCircle(p)[1] for p in contours_poly]

max_index = np.argmax(radii)
radius = int(radii[max_index] * 0.85)
cx, cy = int(centers[max_index][0]), int(centers[max_index][1])

x = cx - radius
y = cy - radius
w = h = 2 * radius
x2 = min(x + w, img.shape[1])
y2 = min(y + h, img.shape[0])

crop_img = img[y:y2, x:x2]



gray_crop = cv.cvtColor(crop_img, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray_crop, (1, 1), 0)

min_radius = 10
max_radius = 30

kernel = cv.getStructuringElement(cv.MORPH_RECT, (3, 3))
morph = cv.morphologyEx(gray_crop, cv.MORPH_GRADIENT, kernel)

circles = cv.HoughCircles(
    gray_crop,
    cv.HOUGH_GRADIENT,
    dp=1,
    minDist=10,
    param1=200,
    param2=45,
    minRadius=min_radius,
    maxRadius=max_radius
)

x1 = 95
y1 = 75
x2 = 95
y2 = 95
cv.circle(crop_img, (x1, y1), 5, (255, 0, 0), -1)
cv.circle(crop_img, (x2, y2), 5, (0, 0, 255), -1)


def find_pointer_line(morph, center_x, center_y, max_r,
                      canny_thresh1, canny_thresh2,
                      hough_threshold=255, minLineLength=20, maxLineGap=20):
    edges = cv.Canny(morph, canny_thresh1, canny_thresh2)
    lines = cv.HoughLinesP(edges, 1, np.pi / 180,
                           threshold=hough_threshold,
                           minLineLength=minLineLength,
                           maxLineGap=maxLineGap)

    best_line = None
    max_length = 0

    if lines is not None:
        for line in lines:
            x1_l, y1_l, x2_l, y2_l = line[0]
            for (px, py) in [(x1_l, y1_l), (x2_l, y2_l)]:
                d = np.hypot(px - center_x, py - center_y)
                if d < max_r * 2:
                    length = np.hypot(x2_l - x1_l, y2_l - y1_l)
                    if length > max_length:
                        best_line = (x1_l, y1_l, x2_l, y2_l)
                        max_length = length
    return best_line

# Стрілка 1 — з одними параметрами
pointer_line1 = find_pointer_line(
    morph, x1, y1, max_radius,
    canny_thresh1=80, canny_thresh2=180,
    hough_threshold=40, minLineLength=60, maxLineGap=5
)
if pointer_line1:
    x1_1, y1_1, x2_1, y2_1 = pointer_line1
    cv.line(crop_img, (x1, y1), (x1_1, y1_1), (0, 255, 0), 2)

blurred = cv.GaussianBlur(morph, (3, 3), 0)

# Стрілка 2 — з іншими параметрами
pointer_line2 = find_pointer_line(
    gray_crop, x2, y2, max_radius,
    canny_thresh1=80, canny_thresh2=180,
    hough_threshold=40, minLineLength=60, maxLineGap=8
)
if pointer_line2:
    x1_2, y1_2, x2_2, y2_2 = pointer_line2
    cv.line(crop_img, (x2, y2), (x1_2, y1_2), (0, 255, 0), 2)

pointer_line3 = find_pointer_line(
    gray_crop, x2, y2, max_radius,
    canny_thresh1=20, canny_thresh2=100,
    hough_threshold=40, minLineLength=60, maxLineGap=8
)
if pointer_line3:
    x1_2, y1_2, x2_3, y2_3 = pointer_line3
    cv.line(crop_img, (x2, y2), (x2_3, y2_3), (0, 255, 0), 2)

def get_pressure(cx, cy, x_end, y_end, scale_max):
    dx = x_end - cx
    dy = y_end - cy  # напрямок вниз (як у OpenCV)

    angle = np.arctan2(dy, dx) * 180 / np.pi  # [-180, 180]
    angle = (angle + 360) % 360  # [0, 360)

    # Переводимо верхню півкулю (від 9 до 3 години) в діапазон 0–180
    angle_from_9 = (angle - 180) % 360

    if angle_from_9 > 180:
        return 0  # не входить у верхню півкулю

    value = int((angle_from_9 / 180) * scale_max)
    return value

def get_temperature(cx, cy, x_end, y_end, scale_max):
    dx = x_end - cx
    dy = y_end - cy 
    angle = np.arctan2(dy, dx) * 180 / np.pi

    angle_from_9 = (180 - angle) % 360

    value = int((angle_from_9 / 180) * scale_max)
    return value

pressure = get_pressure(x1, y1, x1_1, y1_1, scale_max=10)
red_ar = get_pressure(x2, y2, x2_3, y2_3, scale_max=10)
temperature = get_temperature(x2, y2, x2_1, y2_1, scale_max=120)


#cv.putText(crop_img, f"{pressure:.2f}", (0, 30), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.putText(crop_img, str(pressure), (0, 30), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.putText(crop_img, str(temperature), (160, 180), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.putText(crop_img, str(red_ar), (150, 30), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.imshow('contours', crop_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
