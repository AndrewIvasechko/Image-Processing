```python
#пробую новий метод
from __future__ import print_function
import cv2 as cv
import numpy as np
import os


def get_pointer_value(cx, cy, x_end, y_end, scale_max):
    """
    Обчислює значення на круговій шкалі, на яке вказує стрілка.

    cx, cy — координати центра шкали
    x_end, y_end — координати кінця стрілки
    scale_max — максимальне значення шкали (наприклад, 10 або 100)

    Повертає: числове значення
    """
    dx = x_end - cx
    dy = cy - y_end  # інверсія координати Y (бо в OpenCV вона вниз)

    angle = np.degrees(np.arctan2(dy, dx))  # кут у градусах
    angle = (angle + 180) % 180  # 0° = вгору, по годинниковій стрілці

    value = round((angle / 180) * scale_max, 2)
    return value

img = cv.imread('5.jpg')
img = cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_AREA)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold = 150
canny = cv.Canny(gray, threshold, threshold * 2)
contours, _ = cv.findContours(canny, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

contours_poly = [cv.approxPolyDP(c, 3, True) for c in contours]
centers = [cv.minEnclosingCircle(p)[0] for p in contours_poly]
radii = [cv.minEnclosingCircle(p)[1] for p in contours_poly]

max_index = np.argmax(radii)
radius = int(radii[max_index] * 0.85)
cx, cy = int(centers[max_index][0]), int(centers[max_index][1])

x = cx - radius
y = cy - radius
w = h = 2 * radius
x2 = min(x + w, img.shape[1])
y2 = min(y + h, img.shape[0])

crop_img = img[y:y2, x:x2]



gray_crop = cv.cvtColor(crop_img, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray_crop, (1, 1), 0)

min_radius = 10
max_radius = 25

# Пошук кіл методом Хафа
circles = cv.HoughCircles(
    gray_crop,
    cv.HOUGH_GRADIENT,
    dp=1,
    minDist=50,
    param1=100,
    param2=30,
    minRadius=min_radius,
    maxRadius=max_radius
)

# Якщо знайдено хоча б одне коло
if circles is not None:
    circles = np.uint16(np.around(circles))
    for circle in circles[0, :]:
        x, y, r = circle
        #cv.circle(crop_img, (x, y), r, (0, 255, 0), 2)
        #cv.circle(crop_img, (x, y), 2, (255, 0, 0), 3)

        # Пошук стрілки з цього центра
        edges = cv.Canny(gray_crop, 50, 150)
        lines = cv.HoughLinesP(edges, 1, np.pi / 180, threshold=20, minLineLength=30, maxLineGap=5)

        if lines is not None:
            min_dist = float('inf')
            best_line = None

            for line in lines:
                x1, y1, x2, y2 = line[0]
                # Обчислення відстані від центра кола до початку/кінця лінії
                d1 = np.hypot(x1 - x, y1 - y)
                d2 = np.hypot(x2 - x, y2 - y)
                if min(d1, d2) < r * 2:  # лінія проходить через центр
                    length = np.hypot(x2 - x1, y2 - y1)
                    if length > 10 and length < radius * 2:  # уникнути великих контурів циферблату
                        if length > (best_line[4] if best_line else 0):
                            best_line = (x1, y1, x2, y2, length)

            if best_line:
                x1, y1, x2, y2, _ = best_line
                cv.line(crop_img, (x1, y1), (x2, y2), (0, 255, 0), 2)

value = get_pointer_value(cx, cy, x2, y2, scale_max=120)
print("Значення на шкалі:", value)

cv.imshow('contours', crop_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#Почнем заново
from __future__ import print_function
import cv2 as cv
import numpy as np
import os

img = cv.imread('5.jpg')
img = cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_AREA)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold = 150
canny = cv.Canny(gray, threshold, threshold * 2)
contours, _ = cv.findContours(canny, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

contours_poly = [cv.approxPolyDP(c, 3, True) for c in contours]
centers = [cv.minEnclosingCircle(p)[0] for p in contours_poly]
radii = [cv.minEnclosingCircle(p)[1] for p in contours_poly]

max_index = np.argmax(radii)
radius = int(radii[max_index] * 0.85)
cx, cy = int(centers[max_index][0]), int(centers[max_index][1])

x = cx - radius
y = cy - radius
w = h = 2 * radius
x2 = min(x + w, img.shape[1])
y2 = min(y + h, img.shape[0])

crop_img = img[y:y2, x:x2]



gray_crop = cv.cvtColor(crop_img, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray_crop, (1, 1), 0)

min_radius = 10
max_radius = 25

# Пошук кіл методом Хафа
circles = cv.HoughCircles(
    gray_crop,
    cv.HOUGH_GRADIENT,
    dp=1,
    minDist=50,
    param1=100,
    param2=30,
    minRadius=min_radius,
    maxRadius=max_radius
)

if circles is not None:
    circles = np.uint16(np.around(circles))
    
    # Відсортувати за розміром або координатою (наприклад, по y для верхньої/нижньої)
    sorted_circles = sorted(circles[0], key=lambda c: c[1])  # сортуємо за y

    # Отримати координати перших двох
    x1, y1, _ = sorted_circles[0]
    x2, y2, _ = sorted_circles[1]

    # Візуалізація (опціонально)
    cv.circle(crop_img, (x1, y1), 5, (255, 0, 0), -1)
    cv.circle(crop_img, (x2, y2), 5, (0, 0, 255), -1)

def find_pointer_line(gray_img, center_x, center_y, max_r):
    edges = cv.Canny(gray_img, 50, 150)
    lines = cv.HoughLinesP(edges, 1, np.pi / 180, threshold=20, minLineLength=30, maxLineGap=5)

    best_line = None
    max_length = 0

    if lines is not None:
        for line in lines:
            x1_l, y1_l, x2_l, y2_l = line[0]
            for (px, py) in [(x1_l, y1_l), (x2_l, y2_l)]:
                d = np.hypot(px - center_x, py - center_y)
                if d < max_r * 2:
                    length = np.hypot(x2_l - x1_l, y2_l - y1_l)
                    if length > max_length:
                        best_line = (x1_l, y1_l, x2_l, y2_l)
                        max_length = length
    return best_line

# Стрілка 1
pointer_line1 = find_pointer_line(gray_crop, x1, y1, max_radius)
if pointer_line1:
    x1_1, y1_1, x2_1, y2_1 = pointer_line1
    cv.line(crop_img, (x1_1, y1_1), (x2_1, y2_1), (255, 255, 0), 2)

# Стрілка 2
pointer_line2 = find_pointer_line(gray_crop, x2, y2, max_radius)
if pointer_line2:
    x1_2, y1_2, x2_2, y2_2 = pointer_line2
    cv.line(crop_img, (x1_2, y1_2), (x2_2, y2_2), (0, 0, 255), 2)

def get_pressure(cx, cy, x_end, y_end, scale_max):
    dx = x_end - cx
    dy = cy - y_end  # інверсія осі Y для "вгору"

    angle = np.arctan2(dy, dx) * 180 / np.pi  # кут у градусах
    angle = (angle + 360) % 360  # нормалізація в [0, 360)

    # 0° → 9 година (ліворуч), 90° → 12 година (вгору), 180° → 3 година (вправо)
    # Обмежуємо шкалу тільки верхньою половиною (0–180°)
    if angle > 180:
        return None  # стрілка поза шкалою

    value = int((angle / 180) * scale_max)
    return value

def get_temperature(cx, cy, x_end, y_end, scale_max):
    dx = x_end - cx
    dy = y_end - cy  # OpenCV: вісь Y вниз

    angle = np.arctan2(dy, dx) * 180 / np.pi  # [-180, 180]
    #angle = (angle + 360) % 360  # [0, 360)

    # Зсув: 0 на 9 годині → 180 градусів повинно бути 0
    # Враховуємо, що верхня шкала читається ПРОТИ годинникової
    angle_from_9 = (180 - angle) % 360

    # Масштабуємо від 0 (на 9 годині) до 180 (на 3 годині)
    if angle_from_9 > 180:
        angle_from_9 = 360 - angle_from_9

    value = int((angle_from_9 / 180) * scale_max)
    return value


def draw_scale_lines(img, cx, cy, radius=40, angle_shift=0, color=(0, 0, 255)):
    """
    Малює 13 годинних міток від 9 до 3 години навколо заданого центра (cx, cy).
    
    img — зображення (BGR)
    cx, cy — координати центра
    radius — довжина міток
    angle_shift — зсув градусів (0 для шкали від 9 до 3 години, 180 — для зворотної)
    color — колір ліній (за замовчуванням червоний)
    """
    for i in range(13):  # Від 0 до 12
        angle_deg = 180 * i / 12 + angle_shift
        angle_rad = np.deg2rad(angle_deg)
        x_end = int(cx + radius * np.cos(angle_rad))
        y_end = int(cy + radius * np.sin(angle_rad))
        cv.line(img, (cx, cy), (x_end, y_end), color, 1)

pressure = get_pressure(x1_1, y1_1, x1_1, y1_1, scale_max=10)
temperature = get_temperature(x2, y2, x1_2, y1_2, scale_max=120)

draw_scale_lines(crop_img, x1, y1, radius=50, angle_shift=180, color=(255, 255, 0))     # верхня шкала
draw_scale_lines(crop_img, x2, y2, radius=50, angle_shift=0, color=(0, 0, 255))


#cv.putText(crop_img, f"{pressure:.2f}", (0, 30), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.putText(crop_img, str(pressure), (0, 30), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.putText(crop_img, str(temperature), (350, 400), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.imshow('contours', crop_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#final version
from __future__ import print_function
import cv2 as cv
import numpy as np
import os

img = cv.imread('5.jpg')
img = cv.resize(img, None, fx=1.5, fy=1.5, interpolation=cv.INTER_AREA)
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

threshold = 150
canny = cv.Canny(gray, threshold, threshold * 2)
contours, _ = cv.findContours(canny, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)

contours_poly = [cv.approxPolyDP(c, 3, True) for c in contours]
centers = [cv.minEnclosingCircle(p)[0] for p in contours_poly]
radii = [cv.minEnclosingCircle(p)[1] for p in contours_poly]

max_index = np.argmax(radii)
radius = int(radii[max_index] * 0.85)
cx, cy = int(centers[max_index][0]), int(centers[max_index][1])

x = cx - radius
y = cy - radius
w = h = 2 * radius
x2 = min(x + w, img.shape[1])
y2 = min(y + h, img.shape[0])

crop_img = img[y:y2, x:x2]



gray_crop = cv.cvtColor(crop_img, cv.COLOR_BGR2GRAY)
blurred = cv.GaussianBlur(gray_crop, (1, 1), 0)

min_radius = 10
max_radius = 25

circles = cv.HoughCircles(
    gray_crop,
    cv.HOUGH_GRADIENT,
    dp=1,
    minDist=50,
    param1=100,
    param2=30,
    minRadius=min_radius,
    maxRadius=max_radius
)

if circles is not None:
    circles = np.uint16(np.around(circles))
    
    sorted_circles = sorted(circles[0], key=lambda c: c[1])  

    # Отримати координати перших двох
    x1, y1, _ = sorted_circles[0]
    x2, y2, _ = sorted_circles[1]

    # Візуалізація (опціонально)
    #cv.circle(crop_img, (x1, y1), 5, (255, 0, 0), -1)
    #cv.circle(crop_img, (x2, y2), 5, (0, 0, 255), -1)

def find_pointer_line(gray_img, center_x, center_y, max_r):
    edges = cv.Canny(gray_img, 50, 150)
    lines = cv.HoughLinesP(edges, 1, np.pi / 180, threshold=20, minLineLength=30, maxLineGap=5)

    best_line = None
    max_length = 0

    if lines is not None:
        for line in lines:
            x1_l, y1_l, x2_l, y2_l = line[0]
            for (px, py) in [(x1_l, y1_l), (x2_l, y2_l)]:
                d = np.hypot(px - center_x, py - center_y)
                if d < max_r * 2:
                    length = np.hypot(x2_l - x1_l, y2_l - y1_l)
                    if length > max_length:
                        best_line = (x1_l, y1_l, x2_l, y2_l)
                        max_length = length
    return best_line

# Стрілка 1
pointer_line1 = find_pointer_line(gray_crop, x1, y1, max_radius)
if pointer_line1:
    x1_1, y1_1, x2_1, y2_1 = pointer_line1
    cv.line(crop_img, (x1_1, y1_1), (x2_1, y2_1), (0, 255, 0), 2)

# Стрілка 2
pointer_line2 = find_pointer_line(gray_crop, x2, y2, max_radius)
if pointer_line2:
    x1_2, y1_2, x2_2, y2_2 = pointer_line2
    cv.line(crop_img, (x1_2, y1_2), (x2_2, y2_2), (0, 255, 0), 2)

def get_pressure(cx, cy, x_end, y_end, scale_max):
    dx = x_end - cx
    dy = cy - y_end 

    angle = np.arctan2(dy, dx) * 180 / np.pi 
    angle = (angle + 360) % 360  


    if angle > 180:
        return None 

    value = int((angle / 180) * scale_max)
    return value

def get_temperature(cx, cy, x_end, y_end, scale_max):
    dx = x_end - cx
    dy = y_end - cy 

    angle = np.arctan2(dy, dx) * 180 / np.pi

    angle_from_9 = (180 - angle) % 360

    if angle_from_9 > 180:
        angle_from_9 = 360 - angle_from_9

    value = int((angle_from_9 / 180) * scale_max)
    return value

pressure = get_pressure(x1_1, y1_1, x1_1, y1_1, scale_max=10)
temperature = get_temperature(x2, y2, x1_2, y1_2, scale_max=120)

#cv.putText(crop_img, f"{pressure:.2f}", (0, 30), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.putText(crop_img, str(pressure), (0, 30), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.putText(crop_img, str(temperature), (350, 400), cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0), 2, cv.LINE_AA)
cv.imshow('contours', crop_img)
cv.waitKey(0)
cv.destroyAllWindows()
```
