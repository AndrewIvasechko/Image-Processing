```python
import cv2 as cv
import numpy as np

# Завантаження зображень
image = cv.imread('D:/Projects/CV2/Image-Processing/Transformations/sudoku.png')



if image is None or templ is None:
    print("Не вдалося завантажити зображення або шаблон.")
    exit()

# Перетворення зображення у відтінки сірого
gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

templ = cv.resize(gray, (100, 100))

# Розміри шаблону
h, w = templ.shape[:2]

# Ballard
ballard = cv.createGeneralizedHoughBallard()
ballard.setMinDist(10)
ballard.setLevels(360)
ballard.setDp(2)
ballard.setMaxBufferSize(1000)
ballard.setVotesThreshold(40)

ballard.setCannyLowThresh(30)
ballard.setCannyHighThresh(110)
ballard.setTemplate(templ)

positions_ballard = ballard.detect(gray)[0]  # detect повертає кортеж (positions, votes)

# Guil
guil = cv.createGeneralizedHoughGuil()
guil.setMinDist(10)
guil.setLevels(360)
guil.setDp(3)
guil.setMaxBufferSize(1000)

guil.setMinAngle(0)
guil.setMaxAngle(360)
guil.setAngleStep(1)
guil.setAngleThresh(1500)

guil.setMinScale(0.5)
guil.setMaxScale(2.0)
guil.setScaleStep(0.05)
guil.setScaleThresh(50)

guil.setPosThresh(10)

guil.setCannyLowThresh(30)
guil.setCannyHighThresh(110)
guil.setTemplate(templ)

positions_guil = guil.detect(gray)

# Малювання результатів Ballard
if positions_ballard is not None:
    for x, y, scale, angle in positions_ballard:
        rrect = cv.RotatedRect((x, y), (w * scale, h * scale), angle)
        box = cv.boxPoints(rrect)
        box = np.intp(box)
        cv.drawContours(image, [box], 0, (255, 0, 0), 6)

# Малювання результатів Guil
if positions_guil is not None:
    for x, y, scale, angle in positions_guil:
        rrect = cv.RotatedRect((x, y), (w * scale, h * scale), angle)
        box = cv.boxPoints(rrect)
        box = np.intp(box)
        cv.drawContours(image, [box], 0, (0, 255, 0), 2)

# Показати зображення
cv.imshow("result_img", templ)
cv.waitKey(0)
cv.destroyAllWindows()

```
