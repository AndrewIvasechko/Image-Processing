```python
import cv2
import numpy as np
import random

canvas = np.zeros((500, 500, 3), dtype=np.uint8)

n = np.random.randint(25, 50)
n_tr = 5

lines = np.random.randint(0, 500, size=(n, 4))
rectangles = np.random.randint(0, 500, size=(n, 4))
circles = np.random.randint(0, 500, size=(n, 2))
triangles = np.random.randint(0, 600, size=(n_tr, 6))


for (x1, y1, x2, y2) in lines:
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    thickness = np.random.randint(1, 5) 
    cv2.rectangle(canvas,(x1, y1), (x2, y2),color, thickness)

for (x1, y1, x2, y2) in rectangles:
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    thickness = np.random.randint(1, 5) 
    cv2.line(canvas, (x1, y1), (x2, y2), color, thickness)

for (x1, y1) in circles:
    diameter = np.random.randint(50, 150)
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    thickness = np.random.randint(1, 5) 
    cv2.circle(canvas,(x1, y1), diameter, color, thickness)

for (x1, y1, x2, y2, x3, y3) in triangles:
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    pts = np.array([[x1, y1],[x2, y2],[x3, y3]], np.int32)
    pts = pts.reshape((-1,1,2))
    cv2.fillPoly(canvas, [pts], color, 8)


words = ['Test', 'World', 'OpenCV', 'Random', 'Python', 'Magic', 'Computer Vision', 'Canvas']

for _ in range(n):
    text = random.choice(words)
    position = (random.randint(0, 500), random.randint(0, 500))
    color = tuple(np.random.randint(0, 256, size=3).tolist())
    font_scale = random.uniform(0.5, 3.0)
    thickness = random.randint(2, 5)
    font = random.choice([
        cv2.FONT_HERSHEY_SIMPLEX,
        cv2.FONT_HERSHEY_COMPLEX,
        cv2.FONT_HERSHEY_SCRIPT_SIMPLEX
    ])
    cv2.putText(canvas, text, position, font, font_scale, color, thickness)


cv2.imshow('Random Lines', canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
